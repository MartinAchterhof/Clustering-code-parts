```
name = "the Hague"

df = df_denhaag
df = df.drop(df.columns[0], axis=1)
labels = df_denhaag_labels_DBSCAN.cluster
labels = labels.to_numpy()
labels


from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df_scaled = pd.DataFrame(scaler.fit_transform(df))
df_scaled['cluster'] = labels
df_mean = (df_scaled.groupby('cluster').mean())

df_mean = df_mean.rename(columns={0: 'Price', 1: 'Square meters', 2: 'Cubic meters', 3: 'Field size', 4: 'Number of rooms', 5: 'Number of bedrooms', 
                                  6: 'Balcony', 7 : 'Garage', 8: 'Days Online', 9: 'Type Apartment', 10: 'Type Mansion', 11: 'Type Townhouse', 12: 'Type Villa',
                                  13: 'Type Detached house', 14: 'Type Farmhouse', 15: 'No garden', 16: 'A garden'})

df_scaled = df_scaled.rename(columns={0: 'Price', 1: 'Square meters', 2: 'Cubic meters', 3: 'Field size', 4: 'Number of rooms', 5: 'Number of bedrooms', 
                                      6: 'Balcony', 7 : 'Garage', 8: 'Days Online', 9: 'Type Apartment', 10: 'Type Mansion', 11: 'Type Townhouse', 12: 'Type Villa',
                                      13: 'Type Detached house', 14: 'Type Farmhouse', 15: 'No garden', 16: 'A garden'})
```

```
name = "Amsterdam"

df = df_amsterdam
df = df.drop(df.columns[0], axis=1)
labels = df_amsterdam_labels_DBSCAN.cluster
labels = labels.to_numpy()
labels


from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df_scaled = pd.DataFrame(scaler.fit_transform(df))
df_scaled['cluster'] = labels
df_mean = (df_scaled.groupby('cluster').mean())

df_mean = df_mean.rename(columns={0: 'Price', 1: 'Square meters', 2: 'Cubic meters', 3: 'Field size', 4: 'Number of rooms', 5: 'Number of bedrooms', 
                                  6: 'Balcony', 7 : 'Garage', 8: 'Days Online', 9: 'Type Apartment', 10: 'Type Mansion', 11: 'Type Townhouse', 12: 'Type Villa',
                                  13: 'Type Detached house', 14: 'Type Farmhouse', 15: 'No garden', 16: 'A garden'})

df_scaled = df_scaled.rename(columns={0: 'Price', 1: 'Square meters', 2: 'Cubic meters', 3: 'Field size', 4: 'Number of rooms', 5: 'Number of bedrooms', 
                                      6: 'Balcony', 7 : 'Garage', 8: 'Days Online', 9: 'Type Apartment', 10: 'Type Mansion', 11: 'Type Townhouse', 12: 'Type Villa',
                                      13: 'Type Detached house', 14: 'Type Farmhouse', 15: 'No garden', 16: 'A garden'})
```

```
df_mean
```

```
results = pd.DataFrame(columns=['Variable', 'Var'])
for column in df_mean.columns[0:]:
  results.loc[len(results), :] = [column, np.var(df_mean[column])] 

selected_columns = list(results.sort_values('Var', ascending=False,).head(7).Variable.values) + ['cluster']
tidy = df_scaled[selected_columns].melt(id_vars='cluster')

import seaborn as sns

sns.set(rc={'figure.figsize':(14,3)})

ax = sns.barplot(x='cluster', y='value', hue='variable', data=tidy)
ax.legend(loc='upper left')
ax.set_title('City: %s' % name)

plt.show()
```
